### Задание 1

Дан текст (можно взять любой, необязательно этот):
варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг

Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. Биграммы -- это пары стоящих рядом слов, например, "варкалось хливкие", "хливкие шорьки" и т.д.

Подсказка №1: текст нужно разбить на слова с помощью функции split(), а потом использовать цикл for для создания списка биграмм. На каждой итерации цикла нужно печатать биграмму, а потом добавлять ее в список.

Подсказка №2: количество биграмм в тексте = количество слов в тексте - 1.

```python
def find_bigrams(string): #создадим функцию, которая будет находить биграммы
    s = string.split() #в переменную запишем список, который сделаем из строки, которая введется далее
    bigrams = [] # создаем список, где будем хранить биграммы
    for i in range(0, len(s)): # чтобы выделить биграммы, возьмем текущее и следующее слово при помощи индексов
        if i != len(s) - 1: #проверяем условие, не соответствует ли данный индекс последнему элементу списка (за ним ничего нет)
            bigrams.append((s[i], s[i + 1])) #если условие удовлетворено, добавляем кортеж с биграммой в список
    return bigrams
```

```python
# присваиваем текст из Анны Карениной в качестве значения переменной
words = "Все счастливые семьи похожи друг на друга, каждая несчастливая семья несчастлива по-своему. Все смешалось в доме Облонских. Жена узнала, что муж был в связи с бывшею в их доме француженкою-гувернанткой, и объявила мужу, что не может жить с ним в одном доме. Положение это продолжалось уже третий день и мучительно чувствовалось и самими супругами, и всеми членами семьи, и домочадцами"
bigrams = find_bigrams(words) #вызываем функцию, которую создали выше
print('\nСписок с биграммами для данного текста:\n',bigrams)
print('\nПроверим правильное ли кол-во биграмм мы нашли:', len(bigrams) == (len(words.split()) - 1))
```

Подсказка №2: количество биграмм в тексте = количество слов в тексте - 1.

```python
text = '''Удивительная кошка
Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!''' #записываем текст в переменную
print('Длина текста в символах:', len(text)) #узнаем длину текста в символах при помощи len
from string import punctuation #импортируем пуктуацию
print('\nПолучившийся список:\n', text.split()) #создаем из строки список
new_text = [w.strip(punctuation) for w in text.split()] #в новой переменной при помощи функции strip очищаем текст от пунктуации
print('\nСписок слов очищенных от пунктуации:\n', new_text)
united_line = ' '.join(new_text) #объединяем слова при помощи функции join (и пробела) в строку
print('\nПолучившийся текст:\n', united_line)
lowercase = [w if w.islower() else w.lower() for w in new_text] #проверяем, все ли слова в строке написаны в нижнем регистре, если нет, приводим слово к нижнему регистру
print('\nПреобразованный текст:\n', lowercase)
print('Длина текста:', len(lowercase)) #узнаем длину текста в словах
unique_words = set(lowercase) #при помощи множества составляем список уникальных слов
print('Количество уникальных слов:', len(set(lowercase))) #считаем их количество
print('Последние 10 слов:', new_text[-10:]) #распечатываем последние 10 слов, используя индексы (из переменной new_text)
enter_line = '\n'.join(new_text) #преобразуем полученный в пункте 3 список в строку, в которой слова написаны через перенос строки
print('Получившаяся строка:\n', enter_line)
print('Символы с 20 по 30:', enter_line[19:30]) #печатаем символы с 20 по 30, используя индексы и учитывая, что первый символ имеет индекс 0
bigrams = find_bigrams(united_line) #используем функцию для поиска биграмм
print('Получившийся список биграмм:\n', bigrams)
```

```python
def count_frequency(lst_bigrams): #создаем функцию по подсчету частотности биграмм
    counter = dict() #объявим пустой словарь в переменную counter (ключ - биграмма, значение - частотность)
    for i in lst_bigrams:
        counter[i] = lst_bigrams.count(i) #в словарь записываем значения(частоту), соответствующие ключам (самим биграммам)
    return counter
```

```python
frequency = count_frequency(bigrams) #используем функцию, созданную выше
print('Частотность биграмм:\n',frequency)
```
